2. Apply categorical encoding 

```python
categoricals = print(cat.dtypes[cat.dtypes=='object'])
```

```python
encoded_cat = pd.get_dummies(cat, columns=categoricals)
print(encoded_cat.head())
```

we update the categorical dataset after applying the one-hot encoding

```python
cat = encoded_cat
cat
```

we then combine the numerical and categorical dataframe

```python
combined = pd.concat([num, cat], axis=1)
combined
```

```python
combined.dtypes
```

we will do the train-test spliting the model into 'target_B' and 'target_D' to get a more detailes analysis of the model accuracy score

# Target_B as the target

```python
X_train, X_test, y_train, y_test = train_test_split(combined, target_B, test_size=0.2, random_state=42)
```

1. random forest classifier 

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

2. random forest regressor 

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
```

3. model fit and prediction (x_test)

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

4. accuracy score on the classification model 

```python
accuracy = accuracy_score(y_test, y_pred)
print("accuracy_score:", accuracy)
```

5. accuracy score on the regression model 

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
